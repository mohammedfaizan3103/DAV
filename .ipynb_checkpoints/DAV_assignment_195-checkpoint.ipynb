```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Data Loading and Inspections

```python
df = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')  

print("First 5 rows:\n")
df.head()
```

```python
print("\nData types:\n", df.dtypes)
print("\nMissing values:\n", df.isna().sum())
```

## Data Cleaning with numpy/pandas

```python
# Convert User_Score to numeric
df['User_Score'] = pd.to_numeric(df['User_Score'], errors='coerce')

# Replace missing values in numerical columns with median
df['Critic_Score'] = df['Critic_Score'].fillna(np.nanmedian(df['Critic_Score']))
df['User_Score'] = df['User_Score'].fillna(np.nanmedian(df['User_Score']))

# Convert 'Year_of_Release' to numeric
df['Year_of_Release'] = pd.to_numeric(df['Year_of_Release'], errors='coerce')

# Drop rows with critical missing values
df.dropna(subset=['Name', 'Global_Sales'], inplace=True)

# Check cleaned data
df.info()
```

## Data Analysis Operations

### Basic Analysis (Numpy and Pandas)

```python
sales_data = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales']].values

print("Array shape:", sales_data.shape)
print("Data type:", sales_data.dtype)
print("Total global sales (million):", np.sum(sales_data[:, 4]))  
```

#### Array slicing examples

```python
na_sales = sales_data[:, 0]  
first_5_eu_sales = sales_data[:5, 1]  
print("NA Sales example:", na_sales[:5])
print("First 5 EU sales:", first_5_eu_sales)
```

#### Universal functions

```python
normalized_sales = sales_data / np.max(sales_data, axis=0)
print("\nNormalized Sales Data (first 5 rows):")
print(normalized_sales[:5])

rounded_sales = np.round(sales_data[:5], 1)
print("\nRounded Sales Data (1 decimal):")
print(rounded_sales)
```

#### Aggregations

```python
print("\nMean sales by region:")
print("NA:", np.round(df['NA_Sales'].mean(), 3),
      "\nEU:", np.round(df['EU_Sales'].mean(), 3),
      "\nJP:", np.round(df['JP_Sales'].mean(), 3),
      "\nOther:", np.round(df['Other_Sales'].mean(), 3))
```

#### Broadcasting 

```python
regional_weights = np.array([0.4, 0.3, 0.2, 0.1])  
weighted_sales = sales_data[:, :4] * regional_weights 

print("Regional weights:", regional_weights)
print("First 5 rows of original sales data:", sales_data[:5, :4])
print("First 5 rows of weighted sales:", weighted_sales[:5])
```

#### Boolean masking

```python
high_sales_mask = sales_data[:, 4] > 10  # Games with >10M global sales
high_sellers = df[high_sales_mask]
high_sellers[['Name', 'Global_Sales']].sample(10)
```

#### Sorting with argsort

```python
sorted_indices = np.argsort(sales_data[:, 4])[::-1]
top_sellers = df.iloc[sorted_indices[:10]]
print("Top 10 Games by Global Sales (Descending Order):")
top_sellers[['Name', 'Platform', 'Year_of_Release', 'Global_Sales']]
```

#### Partial sorting

```python
top_5_na_indices = np.argpartition(sales_data[:, 0], -5)[-5:]
top_5_na_games = df.iloc[top_5_na_indices].sort_values('NA_Sales', ascending=False)
print("Top 5 Games by North America Sales (Partial Sort):")
top_5_na_games[['Name', 'Platform', 'NA_Sales']]
```

#### Structured Array

```python
filtered_df = df[['Name', 'Platform', 'Year_of_Release', 'Global_Sales']].dropna()

dtype = [('name', 'U50'), ('platform', 'U10'), ('year', 'i4'), ('global_sales', 'f4')]

structured_array = np.array(
    list(zip(
        filtered_df['Name'],
        filtered_df['Platform'],
        filtered_df['Year_of_Release'].astype(int),
        filtered_df['Global_Sales']
    )),
    dtype=dtype
)
print("Structured Array Sample (First 5 Entries):")
print(structured_array[:5])
```

```python
# Sorting structured array by global sales
structured_sorted = np.sort(structured_array, order='global_sales')[::-1]
print("Top 5 Games by Global Sales (Structured Array):")
print(structured_sorted[:5])
```

#### Hierarchial Indexing

```python
multi_index_df = df.set_index(["Genre", "Platform"])

print("=== Hierarchical Index DataFrame (First 10 Rows) ===")
print(multi_index_df[['Name']].head(10))
```

```python
# Get mean sales per (Genre, Platform)
genre_platform_sales = multi_index_df["Global_Sales"].groupby(level=[0, 1]).mean()

genre_platform_sales
```

### Advanced Operations

#### Creating a new column (Pandas)

```python
df['Combined_Score'] = (df['Critic_Score'] * 0.7 + df['User_Score'] * 0.3)  # 70% critic, 30% user weight
print("\nTop 10 Games by Combined Score:")
df.sort_values('Combined_Score', ascending=False)[['Name', 'Critic_Score', 'User_Score', 'Combined_Score']].head(10)
```

####  Concatenation along rows

```python
# Creating a new row as a dictionary
new_game = {'Name': 'Future Game', 'Platform': 'PC', 'Year_of_Release': 2026,
            'Genre': 'Shooter', 'Publisher': 'New Studio', 'NA_Sales': 1.8,
            'EU_Sales': 0.5, 'JP_Sales': 0.3, 'Other_Sales': 0.2, 'Global_Sales': 2.8,
            'Critic_Score': 85, 'Critic_Count': 65, 'User_Score': 8.5, 'User_Count': 300,
            'Developer': 'Future Devs', 'Rating': 'M'}

new_game_df = pd.DataFrame([new_game])

df_appended = pd.concat([df, new_game_df], ignore_index=True)
df_appended.tail(2)
```

#### Concatenation along columns

```python
# Creating extra columns dataset
extra_info = pd.DataFrame({
    'Game_ID': range(1, len(df) + 1),  # Assigning unique IDs
    'Awards_Won': [np.random.randint(0, 5) for _ in range(len(df))]  # Random awards data
})

df_with_extra_info = pd.concat([df, extra_info], axis=1)
df_with_extra_info[['Name', 'Game_ID', 'Awards_Won']].head(3)
```

#### Merging

```python
publisher_details = pd.DataFrame({
    'Publisher': ['Nintendo', 'Sony', 'Microsoft', 'XYZ Games', 'ABC Studios'],
    'Headquarters': ['Japan', 'Japan', 'USA', 'India', 'UK']
})
```

Inner Join

```python
df_merged = pd.merge(df, publisher_details, on='Publisher', how='inner')
df_merged[['Name', 'Publisher', 'Headquarters']].tail(3)
```

Left Join

```python
df_left_join = pd.merge(df, publisher_details, on='Publisher', how='left')
df_left_join[['Name', 'Publisher', 'Headquarters']].tail(3)
```

Right Join

```python
df_right_join = pd.merge(df, publisher_details, on='Publisher', how='right')
df_right_join[['Name', 'Publisher', 'Headquarters']].tail(3)
```

Full Outer Join

```python
df_outer_join = pd.merge(df, publisher_details, on='Publisher', how='outer')
df_outer_join[['Name', 'Publisher', 'Headquarters']].tail(3)
```

#### Aggregation

```python
# Aggregating total global sales by Genre
genre_sales = df.groupby('Genre')['Global_Sales'].sum()
genre_sales.head(10)
```

```python
# Aggregating multiple statistics
agg_stats = df.groupby('Platform').agg({'Global_Sales': ['sum', 'mean', 'max']})
agg_stats.head(10)
```

#### Group By

```python
genre_sales = df.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)
genre_sales.head(5)
```

```python
platform_year_sales = df.groupby(['Platform', 'Year_of_Release'])['Global_Sales'].mean()
platform_year_sales
```

### Pivot Tables

```python
pivot1 = pd.pivot_table(df, 
                        index='Genre', 
                        columns='Rating', 
                        values='Global_Sales', 
                        aggfunc='mean',
                        fill_value=0)
pivot1
```

```python
pivot2 = pd.pivot_table(df,
                            index='Genre',
                            columns='Rating',
                            values='Global_Sales',
                            aggfunc=['mean', 'max', 'sum'],
                            fill_value=0)
pivot2
```

```python
pivot3 = pd.pivot_table(df,
                       index='Genre',
                       values=['Critic_Score', 'User_Score'],
                       aggfunc='mean')
pivot3
```

## Data Visualization

#### Pie Chart

```python
region_sales = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()
plt.figure(figsize=(4, 4))
plt.pie(region_sales, labels=region_sales.index, autopct='%1.1f%%')
plt.title('Market Share by Region')
plt.show()
```

#### Bar Graph

```python
genre_sales2 = df.groupby('Genre')['Global_Sales'].sum()

plt.figure(figsize=(12, 6))
genre_sales2.plot(kind='bar')
plt.title('Total Global Sales by Genre')
plt.ylabel('Sales (in millions)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

#### Line Plot

```python
yearly_sales = df.groupby('Year_of_Release')['Global_Sales'].sum()
plt.figure(figsize=(10, 5))
yearly_sales.plot(kind='line', marker='o')
plt.title('Video Game Sales Trend Over Years')
plt.xlabel('Year')
plt.ylabel('Total Global Sales (in millions)')
plt.grid(True)
plt.show()
```

#### Scatter Plot

```python
df2 = df.sample(500)
top_genres = df2['Genre'].value_counts().nlargest(3).index

df_top_genres = df2[df2['Genre'].isin(top_genres)]

plt.figure(figsize=(6, 4))

# Create scatter plot with only top genres
scatter = sns.scatterplot(
    data=df_top_genres,
    x='Critic_Score',
    y='User_Score',
    hue='Genre',
    palette='bright',  
    alpha=0.8,
    s=80,
    edgecolor='w',
    linewidth=0.5
)

plt.title('Critic vs User Scores (Top 5 Genres)', fontsize=14, pad=15)
plt.xlabel('Critic Score (0-100)', fontsize=12)
plt.ylabel('User Score (0-10)', fontsize=12)

plt.legend(
    title='Genre',
    bbox_to_anchor=(1.05, 0.95),
    loc='upper left',
    frameon=True,
    framealpha=0.9
)

plt.show()
```

#### Box Plot

```python
top_platforms = df['Platform'].value_counts().index[:10]
plt.figure(figsize=(12, 6))
sns.boxplot(data=df[df['Platform'].isin(top_platforms)],
            x='Platform',
            y='Global_Sales')
plt.title('Sales Distribution by Platform (Top 10)')
plt.yscale('log')  # Log scale for better visualization
plt.xticks(rotation=45)
plt.show()
```
